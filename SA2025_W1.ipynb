### 1. What is name of car that has the highest horsepower?  

```python
# 1. Check if index is 'name'
print(cars_df.index.name)

# 2. Ensure horsepower is numeric
cars_df['horsepower'] = pd.to_numeric(cars_df['horsepower'], errors='coerce')

# 3. Find max horsepower value
max_hp = cars_df['horsepower'].max()

# 4. Get all rows with that max horsepower
max_hp_cars = cars_df[cars_df['horsepower'] == max_hp]

# 5. Print index values (car names)
print("Car(s) with highest horsepower:")
for car_name in max_hp_cars.index:
    print(car_name)
```

### 2. How many cars have mpg ≥ 35?

```python
count = (cars_df['mpg'] >= 35).sum()
print(f"Number of cars with mpg ≥ 35: {count}")
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
# Filter the DataFrame
filtered_cars = cars_df[(cars_df['horsepower'] > 100) & (cars_df['weight'] < 3000)]

# Find the most common origin
most_common_origin = filtered_cars['origin'].mode()[0]

print(f"The most common origin for cars with horsepower > 100 and weight < 3000 is: {most_common_origin}")
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
mean_accel = cars_df[cars_df['origin'] == 'japan']['acceleration'].mean()
mean_accel_rounded = round(mean_accel, 2)
print(f"Mean acceleration of cars from Japan: {mean_accel_rounded}")
```

### 5. Which year had the highest average mpg?

```python
# Group by model year and calculate mean mpg
avg_mpg_by_year = cars_df.groupby('model_year')['mpg'].mean()

# Find the year with the highest average mpg
year_highest_avg_mpg = avg_mpg_by_year.idxmax()
highest_avg_mpg = avg_mpg_by_year.max()

print(f"Year with highest average mpg: {year_highest_avg_mpg} (Average mpg: {highest_avg_mpg:.2f})")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Step 1: Calculate median mpg
median_mpg = cars_df['mpg'].median()

# Step 2: Filter cars with mpg above the median
filtered_cars = cars_df[cars_df['mpg'] > median_mpg].copy()

# Step 3: Calculate horsepower to weight ratio
filtered_cars['hp_per_weight'] = filtered_cars['horsepower'] / filtered_cars['weight']

# Step 4: Find the max ratio value
max_ratio = filtered_cars['hp_per_weight'].max()

# Step 5: Get car(s) with the max ratio
best_cars = filtered_cars[filtered_cars['hp_per_weight'] == max_ratio]

# Print results
print("Car(s) with the best horsepower-to-weight ratio and above-median mpg:")
print(best_cars.index.tolist())  # assuming 'name' is the index
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
# Calculate average mpg per year and origin
avg_mpg = cars_df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Plot using Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Average MPG Over the Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# Step 1: Identify car names (index) that appear in more than one model year
name_year_counts = cars_df.groupby(cars_df.index)['model_year'].nunique()
multi_year_cars = name_year_counts[name_year_counts > 1].index

# Step 2: Filter DataFrame for those cars
filtered_cars = cars_df.loc[cars_df.index.isin(multi_year_cars)]

# Step 3: Compute std deviation, count, and mean mpg per car name (index)
stats = filtered_cars.groupby(filtered_cars.index).agg(
    mpg_std=('mpg', 'std'),
    count=('model_year', 'nunique'),
    avg_mpg=('mpg', 'mean')
)

# Step 4: Filter cars with std(mpg) < 1.0 and count >= 2
consistent_cars = stats[(stats['mpg_std'] < 1.0) & (stats['count'] >= 2)]

# Step 5: Sort by count (descending), then avg_mpg (descending)
consistent_cars_sorted = consistent_cars.sort_values(by=['count', 'avg_mpg'], ascending=[False, False])

# Display results
print("Consistent car models (std mpg < 1.0, produced over multiple years):")
print(consistent_cars_sorted[['count', 'avg_mpg']])
```
